```python
import cv2
import numpy as np
import pickle
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img_count = 35
X_found = np.ndarray((img_count,32,32,3), dtype=np.uint8)

for i in range(0,img_count):
    s=str(".\my_data\example_{:05d}.png".format(i+1))
    img = cv2.cvtColor(cv2.imread(s), cv2.COLOR_BGR2RGB)
    X_found[i] = img
```

```python
fig, axs = plt.subplots(5,7, figsize=(15, 6))
fig.subplots_adjust(hspace = .2, wspace=.001)
axs = axs.ravel()
for i in range(len(X_found)-1):
    index = i #random.randint(0, len(found_samples)-1)
    image = X_found[index]
    axs[i].axis('off')
    axs[i].imshow(image)
```
