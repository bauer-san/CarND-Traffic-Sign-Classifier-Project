## Testing to learn what does resize do.
Resize flatens the array, resizes it, then reshapes it.
Using the a=np.resize(z,(5,5)) example:
Original: 10x10
Flatten: 1x100
Resize: 1x25
Reshape: 5x5

```python
x=np.array(range(10))
print(x)
y=np.transpose(x)
z=np.outer(x,y)
print(z)
a=np.resize(z,(5,5))
print(a)
```
